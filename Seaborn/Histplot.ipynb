```python
data="https://raw.githubusercontent.com/Pitsillides91/Python-Tutorials/master/Seaborn%20Tutorial/Marketing%20Raw%20Data.csv"
```

```python
data=pd.read_csv(data)
```

```python
data.head()
```

```python
#Distribution of Revenue
x=data["Revenue"].values
sns.distplot(x)
```

```python
#Distribution of Revenue
sns.histplot(x="Revenue",data=data,kde=True,color="#fcba03")
```

```python
mean=data["Revenue"].mean()
mean
```

```python
sns.distplot(x,color="#03befc")
plt.axvline(mean,0,1,color="red")
# as you can see its a right skewed distribution
# just because the mean is to the right of distribution 
```

```python
a=data["Visitors"].values
mean=data.Visitors.mean()
sns.distplot(a,color="#03fc2c")
# ploting the mean
plt.axvline(mean,color="black")
```
