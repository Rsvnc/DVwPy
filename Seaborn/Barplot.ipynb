```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data="https://raw.githubusercontent.com/Pitsillides91/Python-Tutorials/master/Seaborn%20Tutorial/Marketing%20Raw%20Data.csv"
```

```python
data=pd.read_csv(data)
```

```python
data.head()
```

```python
sns.barplot(x="Month_ID",y="Revenue",data=data) # it takes the mean by default
```

```python
data.groupby("Month_ID").Revenue.mean()
```

```python
# Remove the confidence interval
sns.barplot(x="Month_ID",y="Revenue",data=data,ci=False)
```

```python
# By Promo
sns.barplot(x="Month_ID",y="Revenue",data=data,ci=False, hue="Promo")
```

```python
# Changing direction (orient)
sns.barplot(x="Revenue",y="Month_ID",data=data,ci=False, hue="Promo",orient="h")
```

```python
# Changing direction (orient) !! Dont forget it shows average (mean)
sns.barplot(x="Revenue",y="Year",data=data,ci=False, hue="Promo",orient="h")
```

```python
# Changing color
sns.barplot(x="Revenue",y="Year",data=data,ci=False, hue="Promo",orient="h",color="purple")
```
