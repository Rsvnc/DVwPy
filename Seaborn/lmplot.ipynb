```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data="https://raw.githubusercontent.com/Pitsillides91/Python-Tutorials/master/Seaborn%20Tutorial/Marketing%20Raw%20Data.csv"
```

```python
data=pd.read_csv(data)
```

```python
data.head()
```

```python
# linear regression
sns.lmplot(x="Marketing Spend",y="Revenue",data=data)
```

```python
sns.lmplot(x="Marketing Spend",y="Revenue",data=data,height=10)
```

```python
 # changing color and adding hue
sns.lmplot(x="Marketing Spend", y="Revenue",data=data,hue="Promo",height=10)
```

```python
# remove the confidence interval
sns.lmplot(x="Marketing Spend", y="Revenue",data=data,hue="Promo",height=10,ci=None)
```

```python
# adding markers
sns.lmplot(x="Marketing Spend", y="Revenue",data=data,hue="Promo",height=10,ci=None,markers=["o","x","D"])
```

```python
# Relationship between marketing spend and Revenue by column
sns.lmplot(x="Marketing Spend",y="Revenue",data=data, height=5, col= "Promo")
```

```python
# Relationship between marketing spend and Revenue by column
sns.lmplot(x="Marketing Spend",y="Revenue",data=data, height=5, col= "Promo",
          line_kws={"color":"red"},ci=False)
```

```python
sns.lmplot(x="Marketing Spend",y="Revenue",data=data, height=5, col= "Promo",
          line_kws={"color":"red"},ci=False,
          scatter_kws={"color":"green"})
```

```python
sns.lmplot(x="Marketing Spend",y="Revenue",data=data, height=5, col= "Day_Name",
          line_kws={"color":"red"},ci=False,
          scatter_kws={"color":"green"})
```

```python
# Relationship between marketing spend and Revenue by column by day (col_wrap - add Jitter too (x_jitter)
sns.lmplot(x="Marketing Spend",y="Revenue",data=data, height=5, col= "Day_Name",
          line_kws={"color":"red"},ci=False,
          scatter_kws={"color":"green"},
          col_wrap=3)
```
