```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data="https://raw.githubusercontent.com/Pitsillides91/Python-Tutorials/master/Seaborn%20Tutorial/Marketing%20Raw%20Data.csv"
```

```python
data=pd.read_csv(data)
```

```python
data.head()
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Revenue)
```

```python
data.Revenue.median()
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Day_Name,y=data.Revenue)
```

```python
# horizontal boxplot
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Revenue,y=data.Day_Name)
```

```python
# changing color
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Revenue,y=data.Day_Name,color="#19cbd1")
```

```python
data.head()
```

```python
# adding hue ( You must be define data )
# changing color
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Revenue,y=data.Day_Name, data=data,hue="Promo",color="#19cbd1")
```

```python
# adding hue ( You must be define data )
# changing color
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Day_Name,y=data.Revenue, data=data,hue="Promo",color="#d11984")
```

```python
# where the data  is concentrated and they gain a understanding of distribution
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(x=data.Day_Name,y=data.Revenue,color="#6609b3")
sns.swarmplot(x=data.Day_Name,y=data.Revenue,color="#f7f425")
```
