```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data="https://raw.githubusercontent.com/Pitsillides91/Python-Tutorials/master/Seaborn%20Tutorial/Marketing%20Raw%20Data.csv"
```

```python
df=pd.read_csv(data)
```

```python
df.shape
```

```python
df.head()
```

```python
sns.lineplot(x="Date",y="Revenue",data=df)
```

```python
sns.lineplot(x="Week_ID",y="Revenue",data=df)
```

```python
sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo")
```

```python
sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo",style="Promo") # By promo with style
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo",style="Promo") 
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo",style="Promo",ci=False)
# ci= remove these confidence interval
```

```python
# adding markers
sns.set(rc={"figure.figsize":(12,10)})
sns.lineplot(x="Week_ID",y="Revenue",
             data=df,hue="Promo",
             style="Promo",ci=False,
             markers=True)
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.lineplot(x="Month_ID",y="Revenue",
             data=df,hue="Promo",
             style="Promo",ci=False,
             markers=True)
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.lineplot(x="Month_ID",y="Revenue",
             data=df,hue="Day_Name",
             style="Promo",ci=False,
             markers=True)
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.lineplot(x="Year",y="Revenue",
             data=df,hue="Day_Name",
             style="Promo",ci=False,
             markers=True)
```
