```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset("penguins")
```

```python
df.head()
```

```python
df.shape
```

```python
df=df[df["species"].isin(["Adelie","Gentoo"])]
```

```python
df=df.dropna(how="any")
```

```python
df.shape
```

```python
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex",size="bill_length_mm")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex",size="bill_length_mm",sizes=(10,150))
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex",size="bill_length_mm",sizes=(10,150),alpha=0.5)
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex",size="bill_length_mm",sizes=(10,150),alpha=0.5,style="species")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex",marker="*")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.relplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex",marker="D")
plt.title("Body Mass vs Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
plt.show()
```
