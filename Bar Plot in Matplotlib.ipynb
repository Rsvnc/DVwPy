```python
N=5
menMeans=(20,35,30,35,-27)
womenMeans=(25,32,34,20,-25)
menStd=(2,3,4,1,2)
womenStd=(3,5,2,3,3)
ind=np.arange(N) #the x locations for the groups
width = 0.35 #the width of bars
```

```python
plt.subplots()
p1=plt.bar(ind,menMeans,width,label="Men",color="purple")
p2=plt.bar(ind,womenMeans,width,label="Women",color="yellow")
plt.axhline(0, color='grey', linewidth=0.8)
plt.ylabel("Scores")
plt.title("Scores by group and gender")
plt.xticks(ind,label=["G1","G2","G3","G4","G5"])
plt.legend()
plt.show()
```

```python
df3 = pd.DataFrame(
    {
        'Year': [2016, 2017, 2018, 2019, 2020, 2021],
        'Men': [300, 320, 310, 305, 325, 330],
        'Women': [350, 325, 300, 305, 380, 355],
        'Total': [650,645, 610, 610, 705, 685]
    }
)
```

```python
df3
```

```python
plt.bar(x=df3.Year,height=df3.Total)
plt.show()
```

```python
plt.bar(x=df3.Year,height=df3.Total)
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
plt.bar(x=df3.Year,height=df3.Total,color="grey")
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
colors=["grey","grey","grey","grey","red","grey",] #change the color manually for the highest value of the year
plt.bar(x=df3.Year,height=df3.Total,color=colors)
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
colors=["pink" if  (bar==max(df["Total"])) else "grey" for bar in df.Total]
plt.barh(y=df3.Year,width=df3.Total,color=colors) #horizontaly bar
plt.xlabel("Sales",fontsize=18)
plt.ylabel("Year",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
colors=["purple" if  (bar==max(df["Total"])) else "grey" for bar in df.Total]
plt.bar(x=df3.Year,height=df3.Total,color=colors,width=0.5) #change the width
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
colors=["green" if  (bar==max(df["Total"])) else "grey" for bar in df.Total]
plt.barh(y=df3.Year,width=df3.Total,color=colors,height=0.5) #change the width from horizontally bar
plt.xlabel("Sales",fontsize=18)
plt.ylabel("Year",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
plt.bar(x=df.Year, height=df.Men,color="blue") #compare two different value in the bar
plt.bar(x=df.Year,height=df.Women,color="orange")
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
plt.bar(x=df.Year, height=df.Men,color="blue",width=0.4,label="Men") # Make 2 different values visible
plt.bar(x=df.Year+0.4,height=df.Women,color="orange",width=0.4,label="Women")
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.legend(loc="lower right")
plt.show()
```

```python
plt.bar(x=df.Year, height=df.Men,color="blue") #setting the starting point
plt.bar(x=df.Year,height=df.Women,color="orange",bottom=df.Men)
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```

```python
df["Men Proportion"]=(df["Men"]/df["Total"])*100 #compare their proportions
df["Women Proportion"]=(df["Women"]/df["Total"])*100
plt.bar(x=df.Year, height=df["Men Proportion"],color="blue")
plt.bar(x=df.Year,height=df["Women Proportion"],color="orange",bottom=df["Men Proportion"])
plt.xlabel("Year",fontsize=18)
plt.ylabel("Sales",fontsize=18)
plt.title("Sales over time",fontsize=25)
plt.show()
```
