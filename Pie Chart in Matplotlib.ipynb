```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
exp_values=[1400,600,500,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.pie(exp_values,labels=exp_labels)
```

```python
plt.pie(exp_values,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=2,autopct="%0.2f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=2,autopct="%0.0f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True,explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=2,autopct="%0.0f%%",shadow=True,explode=[0,0.2,0,0,0],startangle=45)
plt.show()
```

```python
labels=["Apple","Samsung","Xiomi","BBk","others"]
share=[10.85,7,6,10,15] #sales in million units
```

```python
plt.pie(x=share,labels=labels)
plt.show()
```

```python
plt.pie(x=share,labels=labels,autopct="%0.2f%%",radius=1.5)
plt.show()
```

```python
plt.pie(x=share,labels=labels,autopct="%0.2f%%",radius=1.5,explode=[0.1,0,0,0,0])
plt.style.use("seaborn-talk")
plt.show() # only explode the 1st slice
```

```python
print(plt.style.available)
```

```python
plt.pie(x=share,labels=labels,autopct="%0.2f%%",radius=1.5,shadow=True,startangle=90)
plt.show()
```

```python
plt.pie(x=share,labels=labels,autopct="%0.2f%%",radius=1.5,shadow=True,startangle=90)
plt.axis("equal")
plt.title("Share of Companies",fontsize=20,color="blue")
plt.legend(loc="upper left")
plt.show()
```
