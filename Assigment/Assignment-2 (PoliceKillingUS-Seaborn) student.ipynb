## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill.head()
```

```python
kill.manner_of_death.unique()
```

**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
plt.figure(figsize=(9,6))
sns.boxplot(x=kill.gender,y=kill.age,hue=kill.manner_of_death,palette="husl");
```

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(9,6))
sns.swarmplot(x=kill.gender,y=kill.age,hue=kill.manner_of_death,palette="Paired");
```

## 2: Plot 12 most dangerous cities

```python
most_dangerous_city=kill.city.value_counts().iloc[:13]
most_dangerous_city
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x=most_dangerous_city.index,y=most_dangerous_city)
plt.xticks(rotation=45)
plt.title("Most dangerous cities",fontdict={"fontsize":18,"fontweight":"bold","color":"purple"});
```

```python
plt.figure(figsize=(9,6))
sns.countplot(x=kill.city, order=most_dangerous_city.index)
plt.xticks(rotation=45)
plt.title("Most dangerous cities",fontdict={"fontsize":18,"fontweight":"bold","color":"purple"});
```

## 3: Plot 20 most dangerous states

```python
kill.head()
```

```python
most_dangerous_state=kill.state.value_counts().iloc[:21]
most_dangerous_state
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x=most_dangerous_state.index, y=most_dangerous_state)
plt.xticks(rotation=45)
plt.title("Most dangerous states",fontdict={"fontsize":18,"fontweight":"bold","color":"purple"});
```

```python
plt.figure(figsize=(9,6))
sns.countplot(x=kill.state,order=most_dangerous_state.index)
plt.xticks(rotation=45)
plt.title("Most dangerous states",fontdict={"fontsize":18,"fontweight":"bold","color":"purple"});
```

## 4: Plot Having mental ilness or not for killed people

```python
having_mental=kill.signs_of_mental_illness.value_counts()
having_mental
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x=having_mental.index,y=having_mental,palette="cubehelix")
plt.xlabel("signs_of_mental_illness")
plt.ylabel("count");
```

## 5: Plot number of Threat Types

```python
threat_types=kill.threat_level.value_counts()
threat_types
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x=threat_types.index,y=threat_types,palette="husl")
plt.xlabel("threat_level")
plt.ylabel("count");
```
